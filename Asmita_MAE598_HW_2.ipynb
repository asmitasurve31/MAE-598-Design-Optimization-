# **Newton's Method :-**

```python
import numpy as np
from numpy import linalg as LA

def func(x):
    f=5*x[0]**2-8*x[0]+12*x[0]*x[1]+10*x[1]**2-14*x[1]+5
    return f

def grad(x):
    g=np.array([10*x[0]-8+12*x[1],12*x[0]+20*x[1]-14])
    return g

def hess(x):
    h=np.array([[10,12],[12,20]])
    return h

def newmet(x0):
        i=0
        c=0
        f=func(x0)
        g=grad(x0)
        h=hess(x0)
        gt=np.transpose(g)
        ng=LA.norm(g)
        while (ng>0.000001):
            f=func(x0)
            g=grad(x0)
            h=hess(x0)
            hi=np.linalg.inv(h)
            gt=np.transpose(g)
            ng=LA.norm(g)
            a=1

            x0=x0-a*(np.dot(hi,g))
            print("iteration{}=(x2,x3)={}".format(i,x0))
            a=(np.dot(gt,g))/(np.dot((np.dot(gt,h)),g))
            if i==100:
                 break
            i=i+1

newmet([1,-1])
```
