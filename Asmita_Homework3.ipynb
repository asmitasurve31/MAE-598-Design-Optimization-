##Generalized Reduced Gradient Algorithm.

```python
import numpy as np
def f(x1,x2,x3):
    return x1**2+x2**2+x3**2
def f_x(x1,x2,x3):
    f_s=np.vstack((2*x2,2*x3))
    f_d=np.array([2*x1],dtype=float)
    return f_s,f_d
def st(x1,x2,x3):
    h1 = x1**2/4+x2**2/5+x3**2/25-1
    h2 = x1+x2-x3
    return h1,h2
def h_x(x1, x2, x3):
    h_s=np.vstack((np.hstack(((2/5)*x2,(2/25)*x3)),np.array([[1,-1]],dtype=float)))
    h_si=np.linalg.inv(h_s)
    h_d=np.vstack(((1/2)*x1,1.))
    return h_si,h_d,h_s
def fd(x1,x2,x3):
    f_s,f_d=f_x(x1, x2, x3)
    h_si,h_d,_=h_x(x1, x2, x3)
    return f_d - np.matmul(f_s.T, np.matmul(h_si, h_d))
def out(x1,x2,x3):
    er=1e-3
    h1,h2=st(x1,x2,x3)
    h=np.vstack((h1,h2))
    h_n=np.linalg.norm(h)
    while h_n>=er:
        dh_i,_,dh=h_x(x1,x2,x3)
        k=1
        ds=np.matmul(dh_i, h)
        x2=x2-ds[0]
        x3=x3-ds[1]
        h1,h2=st(x1, x2, x3)
        h=np.vstack((h1, h2))
        h_n=np.linalg.norm(h)
    return x1,x2,x3, h_n
def l_s(x1,x2,x3):
    a = 1.
    df = fd(x1, x2, x3)
    t = lambda a,x1,x2,x3,df:f(x1,x2,x3)-a*0.3*np.matmul(df,df.T)
    def f_a(x1,x2,x3,a):
        df=fd(x1,x2,x3)
        h_si,h_d,_=h_x(x1, x2, x3)
        x1=x1-a*df.flatten()
        ds=np.matmul(np.matmul(h_si, h_d), df.T).flatten()
        x2=x2+a*ds[0]
        x3=x3+a*ds[1]
        return f(x1,x2,x3)
    while t(a,x1,x2,x3,df)<f_a(x1,x2,x3,a):
        a=0.5*a
        df=fd(x1,x2,x3)
    return a
r=1e-3
x1=np.array([1.0],dtype=float)
x2=np.array([2.0],dtype=float)
x3=np.array([3.0],dtype=float)
i=0
x1,x2,x3,_=out(x1,x2,x3)
f_d_n=np.linalg.norm(fd(x1, x2, x3))
while f_d_n>=r:
    a=l_s(x1, x2, x3)
    x1=x1-a*fd(x1,x2,x3).flatten()
    h_si,h_d,dh=h_x(x1, x2, x3)
    ds=a*np.matmul(np.matmul(h_si, h_d), fd(x1,x2,x3).T).flatten()
    x2=x2+ds[0]
    x3=x3+ds[1]
    x1,x2,x3,_=out(x1,x2,x3)
    f_d_n=np.linalg.norm(fd(x1, x2, x3))
print('x1,x2,x3=({},{},{})'.format(x1,x2,x3))
```

##Differentiable Program (try)

```python
import torch
import torch.optim as optim

# Step 2: Initialize state variables
x1 = torch.tensor(0.0, requires_grad=True)
x2 = torch.tensor(0.0, requires_grad=True)
x3 = torch.tensor(0.0, requires_grad=True)

# Step 3: Define objective and constraints
f = x1**2 + x2**2 + x3**2
h1 = (x1**2 / 4) + (x2**2 / 5) + (x3**2 / 25) - 1
h2 = x1 + x2 - x3

# Step 4: Initialize GRG parameters
k_max = 100
alpha = 0.1

# Optimization loop
optimizer = optim.SGD([x1, x2, x3], lr=alpha)

for k in range(k_max):
    # Step 5.1: Calculate gradients
    f.backward(retain_graph=True)  # Specify retain_graph=True
    h1.backward(retain_graph=True)
    h2.backward(retain_graph=True)

    # Step 5.2: Compute search direction
    d1 = -x1.grad
    d2 = -x2.grad
    d3 = -x3.grad

    # Step 5.3: Update state variables
    x1.data += alpha * d1
    x2.data += alpha * d2
    x3.data += alpha * d3

    # Reset gradients
    x1.grad.zero_()
    x2.grad.zero_()
    x3.grad.zero_()

# Step 7: Get the optimal solution
optimal_x1 = x1.item()
optimal_x2 = x2.item()
optimal_x3 = x3.item()

print("Optimal Solution: x1 =", optimal_x1, "x2 =", optimal_x2, "x3 =", optimal_x3)
```

##Answer:-


###Decision Variables:

###Let $x_{ij}$ be a binary decision variable, where $x_{ij} = 1$ if the truck travels from node $i$ to node $j$, and $x_{ij} = 0$ otherwise.

###Objective Function: Minimize the total cost:

####Minimize: $\sum_{i=0}^{N-1} \sum_{j=0,j≠i}^{N-1} c_{ij} x_{ij}$.

subject to:

####Constraints:

1) Each site must be visited exactly once, excluding the depot:  

$\sum_{i=0}^{N-1} x_{ij}$   for j=1,2,...,N-1

2)The truck must leave and return to the depot:   

$\sum_{i=0}^{N-1} x_{i0} = 1$     to leave the depot<br>
$\sum_{i=0}^{N-1} x_{0i} = 1$     to return the depot

3)Subtour Elimination: To eliminate subtours, we add the following constraints for all subsets $S$ of sites, where $1∈S$:

 $\sum_{i∈S,j∈S,i≠j} x_{ij}≤∣S∣−1$


####Binary Constraints:

$x_{i,j}∈{0,1}$,for all $i,j$

This formulation ensures that the truck visits all sites exactly once and returns to the depot while minimizing the total travel cost.
